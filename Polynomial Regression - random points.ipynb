```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# Polynomial regression

### second order polynomial regression

```python
n = 200
```

```python
X = 5*np.random.rand(n, 1)
X.shape
```

```python
y = 20 + 2.5*X + 1.75*X**2 + 5*np.random.randn(n, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=55)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14, rotation=1)
plt.show()
```

```python
pr = LinearRegression()
```

```python
quadratic = PolynomialFeatures(degree=2)
```

```python
X_quad = quadratic.fit_transform(X_train)
```

```python
pr.fit(X_quad, y_train)
```

```python
X_fit = np.arange(0, 5.1, 0.1)[:, np.newaxis]
X_fit.shape
```

```python
y_fit = pr.predict(quadratic.fit_transform(X_fit))
```

```python
plt.plot(X, y, "b.", label="Training points", markersize=4)
plt.plot(X_fit, y_fit, "r", label="Polynomial fit", linestyle="--")
plt.legend(loc="upper left")
plt.show()
```

```python
y_pred = pr.predict(quadratic.fit_transform(X_test))
```

```python
plt.scatter(y_pred, y_test)
plt.xlabel("Predictions")
plt.ylabel("Real")
plt.show()
```

```python
print(pr.intercept_)
print(pr.coef_)
```

### Third order polynomial regression

```python
n1 = 250
```

```python
X_1 = 7*np.random.rand(n1, 1)
y_1 = 20 + 12*X_1 - 8*X_1**2 + X_1**3 + 6*np.random.randn(n1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size=0.20, random_state=78)
```

```python
plt.plot(X_1, y_1, "b.")
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14, rotation=1)
plt.show()
```

```python
cubic = PolynomialFeatures(degree=3)
```

```python
pr_1 = LinearRegression()
```

```python
X_cubic = cubic.fit_transform(X_train)
```

```python
pr_1.fit(X_cubic, y_train)
```

```python
X_fit_cubic = np.arange(0,7.1, 0.1)[:, np.newaxis]
```

```python
y_fit = pr_1.predict(cubic.fit_transform(X_fit_cubic))
```

```python
plt.plot(X_1, y_1, "b.", label="Training points", markersize=4)
plt.plot(X_fit_cubic, y_fit, "r", label="Polynomial fit", linestyle="-")
plt.legend(loc="upper left")
plt.show()
```

```python
y_pred = pr_1.predict(cubic.fit_transform(X_test))
```

```python
plt.scatter(y_pred, y_test)
plt.xlabel("Predictions")
plt.ylabel("Real values")
plt.show()
```

```python
print(pr_1.intercept_)
print(pr_1.coef_)
```
